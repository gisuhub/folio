```python
import folium

# 지도의 중심 좌표 설정 (예: 서울)
center_coordinates = [37.5665, 126.9780]

# 기본 지도를 생성
my_map = folium.Map(location=center_coordinates, zoom_start=12)

# 마커 추가
folium.Marker(
    [37.5665, 126.9780], 
    popup='Seoul',
    tooltip='Click for more info'
).add_to(my_map)

my_map
```
